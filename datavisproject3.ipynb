```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x=np.arange(1,10,0.1)
y=2*x+5
plt.plot(x,y,color='y')
plt.title('Line plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
# here how to add two plots in the same figure
x=np.arange(1,10,0.1)
y1=2*x+5
y2=3*x+10

plt.subplot(1,2,1)
plt.plot(x,y1,color='r')
plt.title('subplot of x&y1')
plt.subplot(1,2,2)
plt.plot(x,y2,color='g')
plt.title('subplot of x&y2')
plt.show()
```

```python
# bar plot
fruit={'apple':30,'mango':45,'pineapple':10}
names=list(fruit.keys())
quantity=list(fruit.values())
```

```python
names,quantity
```

```python
plt.bar(names,quantity)
```

```python
# customizing bar plot
plt.bar(names,quantity,color='g')
plt.title('data of fruits')
plt.xlabel('names of the fruits')
plt.ylabel('quantity of the fruits')
plt.show()
# used for catogorical values
```

```python
# customizing bar plot
plt.barh(names,quantity,color='r')
plt.title('data of fruits')
plt.xlabel('names of the fruits')
plt.ylabel('quantity of the fruits')
plt.show()
```

```python
# scatter plot
x=[10,20,30,40,50,60,70,80,90]
a=[1,2,3,4,5,4,3,2,9]
b=[5,6,7,8,3,2,1,4,9]
plt.scatter(x,a)
plt.scatter(x,b)
plt.legend(['a','b'])
plt.xlabel('x')
plt.ylabel('a&b')
plt.title('x vs a&b')
plt.show()
```

```python
# scatter plot
x=[10,20,30,40,50,60,70,80,90]
a=[1,2,3,4,5,4,3,2,9]
b=[5,6,7,8,3,2,1,4,9]
plt.scatter(x,a,s=300)
plt.scatter(x,b,s=200,marker='3')
plt.legend(['a','b'])
plt.xlabel('x')
plt.ylabel('a&b')
plt.title('x vs a&b')
plt.show()
```

```python
# creating box plot
one=[1,2,3,4,5,6,7,8,9]
two=[1,2,3,4,5,4,3,2,1]
three=[6,7,8,9,8,7,6,5,4]
plt.boxplot([one,two,three])
plt.show()
# explanation
# it will show starting value of one and ending values of one and median values it show orange color
```

```python
# violin plot
one=[1,2,3,4,5,6,7,8,9]
two=[1,2,3,4,5,4,3,2,1]
three=[6,7,8,9,8,7,6,5,4]
plt.violinplot([one,two,three])
plt.show()
# both boxplot and violin plot are same but diffference is shape of it
```

```python
# violin plot
one=[1,2,3,4,5,6,7,8,9]
two=[1,2,3,4,5,4,3,2,1]
three=[6,7,8,9,8,7,6,5,4]
plt.violinplot([one,two,three],showmedians=True,showmeans=True)
# in above showmedians and showmedians are used showing us to mean and medians are like a line
plt.grid(True)
# above grid is used for a tabel it will shows every point of is like line
plt.title("distribution of data")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
# piechart
fruits=["sugar","rice","curry","water"]
quantity=[29,100,35,150]
plt.pie(quantity,labels=fruits,autopct='%0.1f%%')
plt.show()
```

```python
# piechart
fruits=["sugar","rice","curry","water"]
quantity=[29,100,35,150]
plt.pie(quantity,labels=fruits,autopct='%0.1f%%',startangle=90)
plt.show()
```

```python
# piechart
fruits=["sugar","rice","curry","water"]
quantity=[29,100,35,150]
plt.pie(quantity,labels=fruits,autopct='%0.1f%%',startangle=90,shadow=True,explode=(0,0,0,0.1))
plt.show()
```

```python
# piechart
fruits=["sugar","rice","curry","water"]
quantity=[29,100,35,150]
pie1=plt.pie(quantity,labels=fruits,autopct='%0.1f%%',radius=2,colors=['black','red','green','pink'])
pie2=plt.pie([5],colors='w',radius=1)
# above called donat plot
# above pie2 is used for the giving white radius in the circle.always decrese in the above pie1
plt.show()
```
