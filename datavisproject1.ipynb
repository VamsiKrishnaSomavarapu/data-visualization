```python
sns.scatterplot(x=hospital['humidity'],y=hospital['temperature'])
# and it shows the data in the points
```

```python
sns.lineplot(x=hospital['humidity'],y=hospital['temperature'])
# here it shows graph like lines and lines will show where point is gone high and low
```

```python
# hear iam using barplot same way but it shows two continous varible not happen it gives error
# so i decided to change the column or change datatype of humidity
sns.barplot(x=hospital['clouds_all'],y=hospital['temperature'])
```

```python
hospital['humidity_control']=pd.cut(hospital['humidity'],bins=[0,30,60,100],labels=['low','medium','high'])
# above i changed and cut the humidity and storing the values in the humidity_controls
sns.barplot(x=hospital['humidity_control'],y=hospital['temperature'])

```

```python
sns.displot(hospital['humidity'])
# it shows only single column called displot
```

```python
sns.jointplot(x=hospital['humidity'],y=hospital['temperature'])
# it will join the two variable called the jointplot
```

```python
sns.jointplot(x=hospital['humidity'],y=hospital['temperature'],kind='hex')
# here i used extra like kind='hex' it means hex is like hexadecimal
```

```python
# kde=karnal density estimate
sns.jointplot(x=hospital['humidity'],y=hospital['temperature'],kind='kde')
```

```python
# it is used for the pair vise visuvalisation
sns.pairplot(hospital[['humidity', 'temperature', 'air_pollution_index']])
```
